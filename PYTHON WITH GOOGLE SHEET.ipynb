```python
#following code to import gspread and ServiceAccountCredentials from oauth2client .
#oauth2client:--->>>>>>client.OAuth2 Web Server Flow class is used for both installed and web applications.
#---------------->>>>>>It is created by passing the client ID, client secret, 
#---------------->>>>>>and scope to its constructor: You provide the constructor with a redirect_uri parameter.


import gspread
from oauth2client.service_account import ServiceAccountCredentials
```

```python
# use creds to create a client to in
# use creds to create a client to interact with the Google Drive API
# it provide scope between sheet api and drive
# create a gspread client authorizing it using those credentials.

gdrivegsheet =['https://spreadsheets.google.com/feeds','https://www.googleapis.com/auth/drive']

#credentials

Credential= ServiceAccountCredentials.from_json_keyfile_name('ast.json', gdrivegsheet)
```

```python
# Find a workbook by name and open the first sheet
#access our google sheets call client and pass it the spreadsheet name and getting access to sheet1
client = gspread.authorize(Credential)
```

```python
googlesheet = client.open('astt').sheet1
googlesheet
```

```python
#to get all the values inside the file

ASD = googlesheet.get_all_records()
ASD
```

```python
#to get exact row values in a second row (Since 1st row is the header)

googlesheet.row_values(2)
```

```python
#to get all the column values in the column 'place'

googlesheet.col_values(2)
```

```python
#to extract a particular cell value

#googlesheet.cell(2, 1).value
googlesheet.cell(2, 2).value
```
